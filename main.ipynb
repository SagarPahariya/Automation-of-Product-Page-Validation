```python
import pandas as pd
from bs4 import BeautifulSoup
import selenium
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import xlsxwriter



driver = webdriver.Chrome(ChromeDriverManager().install())
lo_url, lo_name = [], []
in0= input("Do you have excel file : ")   #this input for con. with links or file
if in0 == "N" or  in0 == 'n' :
    in01=input("Enter input one :")    # for input 1
    driver.get(in01)
    in02=input("Enter input two i.e. links of categ. or sub. categ. :")
    driver.get(in02)
    sc0=driver.page_source
    sc1=BeautifulSoup(sc0,'html')
    sc2=sc1.find('ul',{'class':'pagination pull-right'})
    if sc2 != None:
        p0=sc2.find_all('a')
        p01=list(p0)
        p02=(len(p01)-2)
        p03=p01[p02].text
        p04=int(p03)
        for i in range(1,p04+1):
            driver.get(in02+'?page={}'.format(i))
            sc4=driver.page_source
            sc5=BeautifulSoup(sc4,'html')
            sc6 = sc5.findChild()
            sc7 = sc6.find('div', {'class': "content"})
            sc8 = sc7.find('div', {'class': 'products'})
            sc9 = sc8.find('div', {'class': 'row'})
            sc10=list(sc9)

            for i in range(len(sc10)):
                if i % 2 != 0:
                    sc11=sc10[i]
                    sc12 = sc11.find('img')
                    sc13 = sc12.attrs
                    sc14 = sc13.get('src')
                    sc15=sc14.endswith('.png')
                    if sc15 == True :
                        sc16 = sc11.find('a', {'class': 'product-link'})
                        sc17 = sc16.get('href')
                        sc18="https://yoshops.com"+sc17
                        lo_url.append(sc18)
                        sc19 = sc11.find('span', {'class': 'product-title'})
                        sc20 = sc19.text
                        lo_name.append(sc20)


    else :
        driver.get(in02)
        sc4 = driver.page_source
        sc5 = BeautifulSoup(sc4, 'html')
        sc6 = sc5.findChild()
        sc7 = sc6.find('div', {'class': "content"})
        sc8 = sc7.find('div', {'class': 'products'})
        sc9 = sc8.find('div', {'class': 'row'})
        sc10 = list(sc9)

        for i in range(len(sc10)):
            if i % 2 != 0:
                sc11 = sc10[i]
                sc12 = sc11.find('img')
                sc13 = sc12.attrs
                sc14 = sc13.get('src')
                sc15 = sc14.endswith('.png')
                if sc15 == True:
                    sc16 = sc11.find('a', {'class': 'product-link'})
                    sc17 = sc16.get('href')
                    sc18 = "https://yoshops.com" + sc17
                    lo_url.append(sc18)
                    sc19 = sc11.find('span', {'class': 'product-title'})
                    sc20 = sc19.text
                    lo_name.append(sc20)

else :
    in001=input("File name with extension i.e. .csv : ")
    df=pd.read_csv(in001)
    l0=list(df['LineItem Name'])
    l1=list(set(l0))
    driver.get('https://yoshops.com')
    for i in l1:
        elementID=driver.find_element(By.XPATH, '//*[@id="search"]')
        elementID.send_keys(i)
        elementID.submit()
        s = driver.page_source
        src = BeautifulSoup(s, 'html')
        src1 = src.find('div', {'class': 'content'})
        src2=src1.find('div',{'class':'products'})
        if src2 != None :
            src3 = src2.find('div', {'class': 'row'})
            sc10 = list(src3)

            for i in range(len(sc10)):
                if i % 2 != 0:
                    sc11 = sc10[i]
                    sc12 = sc11.find('img')
                    sc13 = sc12.attrs
                    sc14 = sc13.get('src')
                    sc15 = sc14.endswith('.png')
                    if sc15 == True:
                        sc16 = sc11.find('a', {'class': 'product-link'})
                        sc17 = sc16.get('href')
                        sc18 = "https://yoshops.com" + sc17
                        lo_url.append(sc18)
                        sc19 = sc11.find('span', {'class': 'product-title'})
                        sc20 = sc19.text
                        lo_name.append(sc20)


lo_url.insert(0,'Web-url')
lo_name.insert(0,'Name')
workbook = xlsxwriter.Workbook('Output.xlsx')
worksheet = workbook.add_worksheet()
row, col = 0, 0
for j in range(2):
    if j == 0:
        for i in lo_name:
            worksheet.write(row, col, i)
            row += 1
    if j == 1:
        row = 0
        col = j
        for i in lo_url:
            worksheet.write(row, col, i)
            row += 1
workbook.close()


```
